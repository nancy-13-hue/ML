***SCATTER PLOT***

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6])
y=np.array([1,9,4,5,2,4])
z=np.array([2,5,8,7,9,0])
c=["red","hotpink","lightpink","pink","m","lightblue"]
plt.title("scatter plot",fontsize=20)
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x,y,color=c,label="array")
plt.scatter(x,z,color=c,label="array1")
plt.legend()
plt.grid()
plt.show()

```
