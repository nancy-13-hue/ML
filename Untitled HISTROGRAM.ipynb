***HISTOGRAM PLOT***

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x=np.random.randint(1,50,(20))
print(x)
```

```python
y=[19,  1, 24,  9, 11, 15, 16, 14, 38,  7 ,18, 41, 17, 21, 14, 42, 16, 25, 46, 18]
plt.title("Histogram")
plt.xlabel("data")
plt.ylabel("frequency")
l=[5,10,15,20,25,30,35,40,45,50]
plt.hist(y,label="hist",bins=l,color="hotpink",edgecolor="black",linewidth=1)
plt.legend()
plt.show()
```
