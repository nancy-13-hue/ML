***DATA VISUALIZATION MATPLOTLIB LIBRARY***


```python
import matplotlib.pyplot as plt
import numpy as np

x=["BCA","BBA","BA","BA.LLB","B.COM"]
y=[80,50,45,65,95]
z=[95,56,65,23,85]
width=0.3
p=np.arange(len (x))
p1=[j+width for j in p]
plt.xlabel("COURSES")
plt.ylabel("POPULARITY")
plt.title("FUTURE PREDICTION COURSE")

c=["hotpink","blue","purple","lightgreen","yellow"]
plt.bar(p,y,color=c,edgecolor="black",linewidth=1,alpha=0.9,width=0.3)
plt.bar(p1,z,color=c,edgecolor="black",width=0.3)
plt.xticks(p+width/2,x,rotation=20)


plt.show()
```
