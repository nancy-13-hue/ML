***LINE PLOT***

```python
import matplotlib.pyplot as plt
import numpy as np
days=("monday","tuesday","wednesday","thursday","friday")
workers=(50,75,100,150,56)
workers1=(60,100,79,50,65)
plt.title("line plot",fontsize=20)
plt.xlabel("days")
plt.ylabel("numbers")
plt.plot(days,workers,label="2020")
plt.plot(days,workers1,color="pink",label="2022")
plt.grid()
plt.legend()
plt.show()
```
